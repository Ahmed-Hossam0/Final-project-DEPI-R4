##**Creating a dictionary to insert all the created measures in it**

```python
_Measures = {} # creating dictionary to insert mesures in it as key : value => column_name : value
Measures = pd.DataFrame() #the table that we will put the dictionary in it
```

**understanding fact table**

```python
fact_accidents.head()
```

### total injuries measure

```python
_Measures['Total_injuries_num'] = fact_accidents['Total_injuries_num'].sum()
```

### Total vehicles involved mesure

```python

_Measures['Total_vehicles_involved'] = fact_accidents['Number_of_vehicles_involved'].sum()
```

### Total Accidents Mesure

```python

_Measures['Total_Accidents'] = fact_accidents['Date_Key'].count()
```

### Average Vehicles per Accident

```python
_Measures['Average Vehicles per Accident'] = round(_Measures['Total_vehicles_involved'] / _Measures['Total_Accidents'])
```

### Avg Injuries per Accident

```python
_Measures['Avg Injuries per Accident'] = _Measures['Total_injuries_num'] / _Measures['Total_Accidents']
```

### Fatal Injuries

```python
_Measures['Fatal Injuries'] = fact_accidents['Fatal_injuries_num'].sum()
```

### Incapacitating Injuries

```python
_Measures['Incapacitating Injuries'] = fact_accidents['injuries_incapacitating'].sum()
```

### Non-Incapacitating Injuries

```python
_Measures['Non-Incapacitating Injuries'] = fact_accidents['injuries_non_incapacitating'].sum()
```

### Not Visible Injuries

```python
_Measures['Not Visible Injuries'] = fact_accidents['injuries_not_visibily_evident'].sum()
```

### Fatal Injury Rate

```python
_Measures['Fatal Injury Rate'] = _Measures['Fatal Injuries'] / _Measures['Total_injuries_num']
```

### Incapacitating Injury Rate

```python
_Measures['Incapacitating Injury Rate'] = _Measures['Incapacitating Injuries'] / _Measures['Total_injuries_num']
```

Severity Index

```python
_Measures['Severity_Index'] = round(
    (
        fact_accidents['Fatal_injuries_num'].sum()            * 4 +
        fact_accidents['injuries_incapacitating'].sum()       * 3 +
        fact_accidents['injuries_non_incapacitating'].sum()   * 2 +
        fact_accidents['injuries_not_visibily_evident'].sum() * 1
    ) / _Measures['Total_Accidents'], 4
)

print("Severity Index:", _Measures['Severity_Index'])
```

Number of accidents that had at least one fatality

```python
_Measures['Accidents_With_Death_Count'] = int(
    (fact_accidents['Fatal_injuries_num'] > 0).sum()
)
print("Accidents With Death Count:", _Measures['Accidents_With_Death_Count'])
```

Percentage of accidents that had at least one fatality

```python
_Measures['Accidents_With_Death_Rate'] = round(
    _Measures['Accidents_With_Death_Count'] / _Measures['Total_Accidents'], 4
)
print("Accidents With Death Rate:", f"{_Measures['Accidents_With_Death_Rate']:.2%}")
```

## **Page 2 Road analysis**

### **Charts**

Accidents & Severe Injuries by Road Surface

```python

road_summary = (
    df.groupby('roadway_surface_cond')
      .agg(
          Total_Accidents=('Accident_Key', 'count'),
          Severe_Injuries=('injuries_incapacitating', 'sum'),
          Fatal_Injuries=('Fatal_injuries_num', 'sum')
      )
      .reset_index()
)

# Total severe injuries = Fatal + Incapacitating
road_summary['Severe_Injuries'] = (
    road_summary['Severe_Injuries'] +
    road_summary['Fatal_Injuries']
)

# Sort by total accidents
road_summary = road_summary.sort_values(
    'Total_Accidents',
    ascending=False
)

# Convert to long format
road_long = road_summary.melt(
    id_vars='roadway_surface_cond',
    value_vars=['Total_Accidents', 'Severe_Injuries'],
    var_name='Metric',
    value_name='Count'
)

fig = px.bar(
    road_long,
    x='roadway_surface_cond',
    y='Count',
    color='Metric',
    barmode='group',
    text_auto='.2s',
    title='Accidents & Severe Injuries by Road Surface'
)

fig.update_layout(
    xaxis_title='Road Surface',
    yaxis_title='Count'
)

fig.show()
```

Crash Severity by Road Surface

```python

# Count accidents by Road Surface and Crash Severity
road_severity = (
    df.groupby(['roadway_surface_cond', 'Crash_severity'])
      .size()
      .reset_index(name='Total_Accidents')
)

# Sort road surfaces by total accidents
road_order = (
    road_severity.groupby('roadway_surface_cond')['Total_Accidents']
    .sum()
    .sort_values(ascending=False)
    .index
)

# Optional: order severity levels
severity_order = ['Fatal', 'Major', 'Moderate', 'Minor', 'No Injury']

fig = px.bar(
    road_severity,
    x='roadway_surface_cond',
    y='Total_Accidents',
    color='Crash_severity',
    category_orders={
        'roadway_surface_cond': road_order,
        'Crash_severity': severity_order
    },
    title='Crash Severity by Road Surface',
    text_auto='.2s'
)

fig.update_layout(
    xaxis_title='Road Surface',
    yaxis_title='Number of Accidents',
    legend_title='Crash Severity'
)

fig.show()
```

Total Accidents and Total Injuries by Crash Hour

```python
# Aggregate data by crash hour
hourly_summary = (
    df.groupby('crash_hour')
      .agg(
          Total_Accidents=('Accident_Key', 'count'),
          Total_Injuries=('Total_injuries_num', 'sum')
      )
      .reset_index()
      .sort_values('crash_hour')
)

# Convert to long format
hourly_long = hourly_summary.melt(
    id_vars='crash_hour',
    value_vars=['Total_Accidents', 'Total_Injuries'],
    var_name='Metric',
    value_name='Count'
)

# Plot
fig = px.line(
    hourly_long,
    x='crash_hour',
    y='Count',
    color='Metric',
    markers=True,
    title='Total Accidents and Total Injuries by Crash Hour'
)

fig.update_layout(
    xaxis_title='Crash Hour',
    yaxis_title='Count',
    xaxis=dict(dtick=1)
)

fig.show()
```
