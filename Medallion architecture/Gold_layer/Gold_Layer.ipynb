### **Gold Layer - Creating data model using fact and dimension tables**

# **Importing libraries**

```python
import pandas as pd
```

### ***Read Silver layer***

```python
Gold_df = pd.read_csv(
    "/content/sample_data/Silver.csv"
)

Gold_df["crash_date"] = pd.to_datetime(Gold_df["crash_date"])
```

## ***Dimesnion tables***

1.Dim_date

```python
dim_date = pd.DataFrame()

dim_date["Date"] = pd.to_datetime(
    Gold_df["crash_date"].unique()
)

dim_date = dim_date.sort_values("Date")

dim_date["Date_Key"] = range(
    1,
    len(dim_date)+1
)

dim_date["Year"] = dim_date["Date"].dt.year
dim_date["Month"] = dim_date["Date"].dt.month
dim_date["Month_Name"] = dim_date["Date"].dt.month_name()
dim_date["Day"] = dim_date["Date"].dt.day
dim_date["Quarter"] = dim_date["Date"].dt.quarter
dim_date["Day_Of_Week"] = dim_date["Date"].dt.day_name()
```

2.Dim_Weather

```python
dim_weather = pd.DataFrame({
    "Weather": Gold_df["Weather"].dropna().unique()
})

dim_weather["Weather_Key"] = range(
    1,
    len(dim_weather)+1
)
```

3.Dim_Lighting

```python
dim_lighting = pd.DataFrame({
    "Lighting": Gold_df["Lighting"].dropna().unique()
})

dim_lighting["Lighting_Key"] = range(
    1,
    len(dim_lighting)+1
)
```

4.Dim_Crash_Type

```python
dim_crash_type = Gold_df[[
    "crash_type",
    "Collision Type",
    "Intersection_accident?"
]].dropna().drop_duplicates()

dim_crash_type["Crash_Type_Key"] = range(
    1,
    len(dim_crash_type)+1
)
```

5.Dim_Road

```python
dim_road = Gold_df[[
    "roadway_surface_cond",
    "Road_alignment",
    "trafficway_type",
    "road_defect",
    ]].dropna().drop_duplicates()

dim_road["Road_Key"] = range(
    1,
    len(dim_road) + 1
    )
```

6.Dim_Time

```python
dim_time = Gold_df[[
    "crash_hour",
    "AM/PM",
    "Time Buckets"
]].drop_duplicates()

dim_time["Time_Key"] = range(
    1,
    len(dim_time)+1
)
```

7.Dim_damage

```python
dim_damage = Gold_df[[
    "Cost_of_damage",
    "Damage_level"
]].drop_duplicates()

dim_damage["Damage_Key"] = range(
    1,
    len(dim_damage)+1
)
```

8.Dim_Severity

```python
dim_severity = Gold_df[[
    "Crash_severity"
]].drop_duplicates()

dim_severity["Severity_Key"] = range(
    1,
    len(dim_severity)+1
)
```

### **Fact table**

Merging keys to create relationships

```python
fact = Gold_df.copy()

fact = fact.merge(
    dim_date[["Date_Key","Date"]],
    left_on="crash_date",
    right_on="Date",
    how="left"
)

fact = fact.merge(
    dim_weather,
    on="Weather",
    how="left"
)

fact = fact.merge(
    dim_lighting,
    on="Lighting",
    how="left"
)

fact = fact.merge(
    dim_crash_type,
    on=[
          "crash_type",
          "Collision Type",
          "Intersection_accident?",
    ],
    how="left"
)

fact = fact.merge(
    dim_road,
    on=[
         "roadway_surface_cond",
          "Road_alignment",
          "trafficway_type",
          "road_defect",
    ],
    how="left"
)

fact = fact.merge(
    dim_time,
    on=[
        "crash_hour",
        "AM/PM",
        "Time Buckets"
    ],
    how="left"
)

fact = fact.merge(
    dim_damage,
    on=[
        "Cost_of_damage",
        "Damage_level"
    ],
    how="left"
)

fact = fact.merge(
    dim_severity,
    on=[
        "Crash_severity"
    ],
    how="left"
)
```

Saving keys and numeric columns in fact table

```python
fact_accidents = fact[[
    "Date_Key",
    "Weather_Key",
    "Lighting_Key",
    "Crash_Type_Key",
    "Road_Key",
    "Time_Key",
    "Damage_Key",
    "Severity_Key",

    "Number_of_vehicles_involved",
    "Total_injuries_num",
    "Fatal_injuries_num",
    "injuries_incapacitating",
    "injuries_non_incapacitating",
    "injuries_not_visibily_evident"
]]
```

Adding accident_key for fact table

```python
fact_accidents.insert(
    0,
    "Accident_Key",
    range(1, len(fact_accidents)+1)
)
```

## **Saving all tables as csv files**

```python
# Saving dim_date table
dim_date.to_csv(
    "/content/sample_data/dim_date.csv",
    index=False
)

# Saving dim_weather table
dim_weather.to_csv(
    "/content/sample_data/dim_weather.csv",
    index=False
)

# Saving dim_lighting table
dim_lighting.to_csv(
    "/content/sample_data/dim_lighting.csv",
    index=False
)

# Saving dim_crash_type table
dim_crash_type.to_csv(
    "/content/sample_data/dim_crash_type.csv",
    index=False
)

# Saving dim_road table
dim_road.to_csv(
    "/content/sample_data/dim_road.csv",
    index=False
)

# Saving dim_time table
dim_time.to_csv(
    "/content/sample_data/dim_time.csv",
    index=False
)

# Saving dim_damage table
dim_damage.to_csv(
    "/content/sample_data/dim_damage.csv",
    index=False
)

# Saving dim_severity table
dim_severity.to_csv(
    "/content/sample_data/dim_severity.csv",
    index=False
)

# Saving fact_accidents table
fact_accidents.to_csv(
    "/content/sample_data/fact_accidents.csv",
    index=False
)
```
